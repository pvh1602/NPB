### Final Figure

##### Sparsity 99%

```python

node = np.log([646,147, 402, 192])
path = np.log10([1.59e17, 1.11e29,1.06e23, 6.33e26])
c = [1,2,3,4]
labels = ['PHEW', 'SynFlow', 'RareGem', 'Ours']
acc = np.array([70.44, 66.05, 72.22, 67.17])
markers = ['o', 'v', '*', 's']
acc_scale = np.array([30,10,40,20])
# acc_scale = 

fig, ax = plt.subplots()
fig.set_size_inches(4.5,3.5)

scatter = ax.scatter(node, path, c=c, s=acc_scale*10)

from matplotlib.legend import Legend
# Sort the label
leg = Legend(ax, scatter.legend_elements()[0],['PHEW', 'SynFlow', 'RareGem', 'Ours'],
             loc="lower left", title="Method")
ax.add_artist(leg);

# produce a legend with a cross section of sizes from the scatter
handles, labels = scatter.legend_elements(prop="sizes", alpha=0.6)
legend2 = ax.legend(handles, np.sort(acc), loc="upper right", title="Accuracy (%)")

plt.xlabel('Number of effective nodes (ln scale)')
plt.ylabel('Number of effective paths (log scale)')
plt.title('Sparsity 99%')
plt.tight_layout()
plt.savefig('./RareGem_cifar10_resnet20/compare_with_raregem_s_99.pdf')
plt.show()
```
